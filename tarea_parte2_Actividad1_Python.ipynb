## 1. Elección del Conjunto de Datos

Descripción:
Este conjunto de datos contiene información sobre la calidad del aire en diversas ciudades de EE.UU. entre 2000 y 2016, incluyendo la concentración promedio de contaminantes como NO2, O3, SO2 y CO.

## 2. Limpieza y Transformación de los Datos

```python
#ver valores faltantes
df.isnull().sum()
```

```python
df["Date Local"] = pd.to_datetime(df["Date Local"])
```

```python
df = df.dropna()
```

```python
# Renombrar columnas para facilidad de uso
df = df.rename(columns={
    'NO2 Mean': 'NO2',
    'O3 Mean': 'O3',
    'SO2 Mean': 'SO2',
    'CO Mean': 'CO'
})
```

```python
# Verificar valores duplicados
df = df.drop_duplicates()
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df.head()
```

## 3. Análisis exploratorio de datos (EDA)

```python
from ydata_profiling import ProfileReport
report = ProfileReport(df)
report.to_file('report.html')
```

```python
# Asegúrate de tener la columna Year
df["Year"] = df["Date Local"].dt.year

# Contaminación promedio por año
df_yearly = df.groupby("Year")[["NO2", "O3", "SO2", "CO"]].mean()
print("Contaminación promedio por año:")
print(df_yearly)

# Top 10 ciudades con mayor NO2 promedio
df_city_avg = df.groupby("City")[["NO2", "O3", "SO2", "CO"]].mean()
df_city_avg = df_city_avg.sort_values(by="NO2", ascending=False).head(10)
print("\nTop 10 ciudades con mayor NO2 promedio:")
print(df_city_avg)
```

## 4. Visualización de Resultados


```python
import matplotlib.pyplot as plt
import seaborn as sns

#estilo
sns.set(style="whitegrid")
```

```python
# Evolución de contaminantes a lo largo del tiempo
plt.figure(figsize=(10,6))
df_yearly.plot(marker='o')
plt.title("Contaminantes Promedio por Año (2000–2016)")
plt.ylabel("Concentración Promedio")
plt.xlabel("Año")
plt.legend(["NO2", "O3", "SO2", "CO"])
plt.grid(True)
plt.show()
```

```python
# Gráfico de calor (correlación entre contaminantes)
plt.figure(figsize=(6,5))
sns.heatmap(df[["NO2", "O3", "SO2", "CO"]].corr(), annot=True, cmap="YlGnBu")
plt.title("Correlación entre contaminantes")
plt.show()
```

```python
# Gráfico de barras: Ciudades con más NO2 promedio
plt.figure(figsize=(10,6))
sns.barplot(x=df_city_avg["NO2"], y=df_city_avg.index)
plt.title("Top 10 Ciudades con Mayor NO2 Promedio")
plt.xlabel("NO2 Promedio")
plt.ylabel("Ciudad")
plt.show()
```
